# Recognition of TrySail

学習したモデルを用いて認識を行う

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
from keras.models import load_model
```

```python
model = load_model('./data/1117_cnn_model.h5')
```

```python
image = cv2.imread("./data/momo_asakura/momo_asakura1_0.jpg")
b,g,r = cv2.split(image)
image = cv2.merge([r,g,b])
```

```python
#予測
def detect_who(img):
    name=""
    print(model.predict(img))
    nameNumLabel = np.argmax(model.predict(img))
    
    if nameNumLabel == 0: 
        name="Shina"
    elif nameNumLabel == 1:
        name="Momo"
    elif nameNumLabel == 2:
        name="Sora"
    return name
```

```python
def detect_face(image):
    print(image.shape)
    #グレイスケール化
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    cascade = cv2.CascadeClassifier("./data/haarcascade_frontalface_alt2.xml")
    # 顔認識の実行
    face = cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=2, minSize=(64,64))
    #顔が１つ以上検出された時
    if len(face) > 0:
        for rect in face:
            x,y,width,height = rect
            cv2.rectangle(image, tuple(rect[0:2]), tuple(rect[0:2]+rect[2:4]), (255, 0, 0), thickness=3)
            img = image[rect[1]:rect[1]+rect[3],rect[0]:rect[0]+rect[2]]
            if image.shape[0]<64:
                print("too small")
                continue
            img = cv2.resize(image,(64,64))
            img = np.expand_dims(img,axis=0)
            name = detect_who(img)
            cv2.putText(image, name, (x,y+height+20), cv2.FONT_HERSHEY_DUPLEX,1,(255,0,0),2)
    #顔が検出されなかった時
    else:
        print("no face")
    return image
```

```python
who = detect_face(image)
```

```python
plt.imshow(who)
plt.show()
```
